```python
df = pd.DataFrame([[1,      np.nan, 2],
                   [2,      3,      5],
                   [np.nan, 4,      6]])
df
```

```python
df.dropna()
```

```python
df.dropna(axis='columns')
```

```python
df[3] = np.nan
df
```

```python
df.dropna(axis='columns', how='all')
```

```python
df.dropna(axis='rows', thresh=3)
```

```python
data.fillna(0)
```

```python
# forward-fill
data.fillna(method='ffill')
```

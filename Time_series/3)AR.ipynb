```r
set.seed(1)
x <- w <- rnorm(100)
for (t in 2:100) x[t] <- 0.6*x[t-1] + w[t]
```

```r
layout(1:2)
plot(x, type="l")
acf(x)
```

```r
x.ar <- ar(x, method = "mle")
x.ar$order
```

```r
x.ar$ar
```

```r
x.ar$ar + c(-1.96, 1.96)*sqrt(x.ar$asy.var)
```

```r
require(quantmod)
getSymbols("AMZN")
```

```r
plot(Cl(AMZN))
```

```r
amznrt = diff(log(Cl(AMZN)))
```

```r
plot(amznrt)
```

```r
acf(amznrt, na.action=na.omit)
```

```r
amznrt.ar <- ar(amznrt, na.action=na.omit)
amznrt.ar$order
```

```r
amznrt.ar$ar
```

```r
cov_amz = amznrt.ar$asy.var
cov_amz
```

```r
-0.0278 + c(-1.96, 1.96)*sqrt(cov_amz[1,1])
```

```r
-0.0687 + c(-1.96, 1.96)*sqrt(cov_amz[1,1])
```
